## IQR method to detect outliers

```python
#IQR method
q1 = df.quantile(0.25)
```

```python
q3 = df.quantile(0.75)
```

```python
IQR = q3-q1
```

```python
LL = q1-(1.5*IQR)
```

```python
UL = q3+(1.5*IQR)
```

```python
dff = df[(df>LL)&(df<UL)]
```

```python
dff.isna().sum()
```

```python
dff.dropna(subset='study_hours_per_day', inplace = True)
```

```python
dff.dropna(subset='social_media_hours', inplace = True)
```

```python
dff.dropna(subset='netflix_hours', inplace = True)
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_train, Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_predict = pd.DataFrame(model.predict(X_test), columns = ['y_predicted'])
```

```python
y_predict
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(Y_test, y_predict)
```
